You may erroneously expect list1 to be equal to [10] and list3 to match with [‘a’], thinking that the list
argument will initialize to its default value of [] every time there is a call to the extendList.

However, the flow is like that a new list gets created once **after the function is defined**. 
And the same getsreused whenever someone calls the extendList method without a list argument. 

This is because the calculation of expressions (in default arguments) occurs **at the time of function definition**, not during its invocation.

The list1 and list3 are hence operating on the same default list, whereas list2 is running on a separate object that it has created on its own (by passing an empty list as the value of the list parameter).

```python
del extendList
def extendList(val, list=[]):
    list.append(val)
    return list
```

```python
list1 = extendList(10)
print (f"list1 = {list1}")
list2 = extendList(123,[])
print (f"list2 = {list2}")
list3 = extendList('a')
print (f"list3 = {list3}")
```

## Closure

Closures are functions that "remember" variables from their enclosing scope, even after that scope has finished executing.

```python
def make_multiplier(n):
    print(f" n is: {n}")
    def multiply(x):
        print(f"x: is {x} n is: {n}")
        return x * n
    return multiply
```

```python
    
times_3 = make_multiplier(3)
```

```python
times_3(10)  # 30
```

```python
def create_handler(prefix):
    def handler(event):
        print(f"{prefix}: {event}")
    return handler
```

```python
error_handler = create_handler("ERROR")
warning_handler = create_handler("WARNING")
error_handler("File not found")  # ERROR: File not found
warning_handler("File not found")  # ERROR: File not found
```
